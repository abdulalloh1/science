### Задание 1 (8 баллов):

Загрузите данные под названием «time_series» и постройте график свойств данных временных рядов (тренд, сезонность, цикличность).  

```python
# ваш код программы
import pandas as pd

# Load the data from the provided CSV file
file_path = 'Electricity_production.csv'
time_series_data = pd.read_csv(file_path)

# Display the first few rows of the data to understand its structure
time_series_data.head()
```

```python
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Convert the DATE column to datetime
time_series_data['DATE'] = pd.to_datetime(time_series_data['DATE'])

# Set the DATE column as the index
time_series_data.set_index('DATE', inplace=True)

# Perform seasonal decomposition
result = seasonal_decompose(time_series_data['elektr_ishlab_chiqarish'], model='multiplicative', period=12)

# Plot the decomposition
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 8))
result.observed.plot(ax=ax1, title='Original')
result.trend.plot(ax=ax2, title='Trend')
result.seasonal.plot(ax=ax3, title='Seasonality')
result.resid.plot(ax=ax4, title='Residuals')
plt.tight_layout()
plt.show()
```

### Задание 2 (10 баллов):

a) Проверьте данные на стационарность и измените нестационарные данные на стационарные данные.

b) Проверьте стационарность, используя методы скользящего среднего и стандартного отклонения, и нарисуйте результаты графически.


```python
from statsmodels.tsa.stattools import adfuller
import numpy as np

# Function to perform the ADF test
def adf_test(timeseries):
    result = adfuller(timeseries, autolag='AIC')
    adf_stat = result[0]
    p_value = result[1]
    return adf_stat, p_value

# Perform ADF test on the original data
adf_stat, p_value = adf_test(time_series_data['elektr_ishlab_chiqarish'])

rolling_mean = time_series_data['elektr_ishlab_chiqarish'].rolling(window=12).mean()
rolling_std = time_series_data['elektr_ishlab_chiqarish'].rolling(window=12).std()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(time_series_data['elektr_ishlab_chiqarish'], label='Original')
ax.plot(rolling_mean, label='Rolling Mean')
ax.plot(rolling_std, label='Rolling STD', color='red')
ax.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation')
plt.show()

adf_stat, p_value, rolling_mean.tail(), rolling_std.tail()
```

```python
time_series_diff = time_series_data['elektr_ishlab_chiqarish'].diff().dropna()

adf_stat_diff, p_value_diff = adf_test(time_series_diff)

rolling_mean_diff = time_series_diff.rolling(window=12).mean()
rolling_std_diff = time_series_diff.rolling(window=12).std()

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(time_series_diff, label='Differenced Data')
ax.plot(rolling_mean_diff, label='Rolling Mean of Differenced Data')
ax.plot(rolling_std_diff, label='Rolling STD of Differenced Data', color='red')
ax.legend(loc='best')
plt.title('Differenced Data with Rolling Mean & Standard Deviation')
plt.show()

adf_stat_diff, p_value_diff
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

train_data, test_data = train_test_split(time_series_data['elektr_ishlab_chiqarish'], test_size=0.2, shuffle=False)

scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_data.values.reshape(-1, 1))
test_scaled = scaler.transform(test_data.values.reshape(-1, 1))

train_scaled = pd.DataFrame(train_scaled, index=train_data.index, columns=['Scaled Data'])
test_scaled = pd.DataFrame(test_scaled, index=test_data.index, columns=['Scaled Data'])

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(train_scaled, label='Scaled Training Data')
ax.plot(test_scaled, label='Scaled Testing Data', color='red')
ax.legend(loc='best')
plt.title('Scaled Training and Testing Data')
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

plot_acf(train_data, lags=40, ax=ax1)
ax1.set_title('Autocorrelation Function')

plot_pacf(train_data, lags=40, ax=ax2, method='ywm')
ax2.set_title('Partial Autocorrelation Function')

plt.tight_layout()
plt.show()
```
